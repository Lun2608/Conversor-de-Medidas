### -----------------------------CONVERSOR DE MEDIDAS-----------------------------

O projeto se trata de uma conversor de medidas, com interface gráfica, no qual o usuário escolhe o que gostaria de converter, podendo escolher entre temperatura, unidades de distância e moedas. Mais adiante estarão mais explicações sobre o funcionamento do código.

### Limitações do código

Antes de apresentar o código, deve-se ressaltar que ele apresenta limitações em relação as próprias conversões. A seguir cito algum dos problemas que podem ocorrer durante o uso do programa:

<ul>
    <li>As transformações são arrendondas, perdendo sua precisão exata;</li>
    <li>As unidades para as conversões foram previamente selecionadas, não contendo todas as unidades existentes;</li>
    <li>O resultado é apresentado com apenas duas casas decimais, de modo que números muito pequenos se apresentam como zero;</li>
    <li>As transformações no Conversor de Moedas são fixas (valores aproximados referentes a junho de 2026), de modo que podem não representar com precisão as cotações atuais;</li>
    <li>Se o usuário submeter letras, o código não irá funcionar;</li>
    <li> Se o usuário submeter números muito grandes, o código não irá funcionar.</li>

### Interface gráfica

A interface foi construída para oferecer uma experiência de uso simples e direta. O usuário insere o valor que deseja converter, seleciona a unidade de entrada e a unidade de saída por meio de botões, e obtém o resultado de forma imediata. Essa abordagem elimina a necessidade de interação via linha de comando, tornando o programa acessível a pessoas sem familiaridade com programação.

### Começando o código

### Definindo as funções

</ul>
